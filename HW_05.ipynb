Исходные данные имеют следующую структуру:

1. перечень всех документов
```

```


Общие требования к программе:

- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
# Задаем наши переменные

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
def command_p():
    user_unput = input('Введите номер документа: ')
    print('')
    for document in documents:
        if document['number'] == user_unput:
            return print('Владелец документа: ', document['name'])
    return print('Документ не найден в базе')
````

````python
def command_s():
    user_unput = input('Введите номер документа: ')
    print('')
    for a, b in directories.items():
        for c in b:
            if c == user_unput:
                return print('Документ хранится на полке: ',a)
    return print('Документ не найден в базе')
````

````python
def command_l():
    for document in documents:
        for a, b in directories.items():
            for c in b:
                if c == document['number']:
                    print('№: ', document['number'],', тип: ', document['type'], ', владелец: ', document['name'], ', полка хранения: ', a)
````

````python
def directories_print():
    return ', '.join(directories.keys())
````

````python
def command_ads():
    user_unput = input('Введите номер полки: ')
    print('')
    for a in directories.keys():
        if a == user_unput:
            return print('Такая полка уже существует. Текущий перечень полок: ', directories_print())
    directories[str(user_unput)] = []
    return print('Полка добавлена. Текущий перечень полок: ', directories_print())
````

````python
def command_ds():
    user_unput = input('Введите номер полки: ')
    print('')
    for a, b in directories.items():
        if a == user_unput:
            if len(b) != 0:
                return print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', directories_print())
            else:
                del directories[user_unput]
                return print('Полка удалена. Текущий перечень полок: ', directories_print())
    return print('Такой полки не существует. Текущий перечень полок: ', directories_print())
````

````python
def main():
    '''
    Функция отвечающая за ввод команды:
    
    Поддерживаемые команды:
    
    p - поиск владельца по номеру документа
    s - поиск полки по номеру документа
    l - полная информация по всем документам
    ads - добавляет новую полку
    ds - удаление пустой полки
    q - завершить программу
    '''
    while True:
        user_unput = input('Введите команду: ')
        print('')
        if user_unput == 'q':
            print('Работа программы завершена')
            break 
        elif user_unput == 'p':
            command_p(), print('')
        elif user_unput == 's':
            command_s(), print('')
        elif user_unput == 'l':
            command_l(), print('')
        elif user_unput == 'ads':
            command_ads(), print('')
        elif user_unput == 'ds':
            command_ds(), print('')
        else:
            print('Неизвестная команда, повторите ввод!'), print('')
````

````python
main()
````
