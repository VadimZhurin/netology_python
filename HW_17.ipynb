Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

- 777: 200 руб.
- 999: 100 руб.
- 555: 50 руб.
- 333: 15 руб.
- 111: 10 руб.
- *77: 5 руб.
- **7: 3 руб.
- *00: 2 руб.
- **0: 1 руб.

Вопросы:

- Выгодна ли игра игроку?
- Сколько в среднем приобретает или теряет игрок за одну игру?

**Дополнительное задание повышенной сложности.**

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

- любая цифра

```python
import numpy as np
```

```python
games = 1000000 # количество игр / сумма проигранных денег
wins = 0 # количество выигрышей
total = 0 # сумма выигранных денег
wins_combs1 = [[777,200],[999,100],[555,50],[333,15],[111,10]] # список абсолютных выигрышных чисел
wins_combs2 = [[100,2]] # если число заканчивается на несколько нулей
wins_combs3 = [[77,5],[7,3],[0,1]] # список окончаний чисел

for i in range(games):
    combs = np.random.randint(0, 1000) # случайная комбинация чисел 0 до 999
    stop = 0 # это будет наш счетчик для остановки циклов
    for a, b in wins_combs1: 
        if combs == a: # проверяем совпадает ли наше число с абсолютно выигрышным, если да то:
            wins += 1 # фиксируем 1 победу
            total += b # увеличиваем наш выигрыш
            stop += 1 # фиксируем выигрыш увеличивая счетчик
            break # завершаем цикл 
    if stop == 0: # если счетчик не изменился, ищем выигрыш в следующем шаге
        for c, d in wins_combs2: 
            if combs / c % 1 == 0: # проверяем делиться ли наше число на число имеющее больше одного нуля в конце целиком:
                wins += 1 # фиксируем 1 победу
                total += d # увеличиваем наш выигрыш
                stop += 1 # фиксируем выигрыш увеличивая счетчик
                break # завершаем цикл
    if stop == 0: # если счетчик не изменился, ищем выигрыш в следующем шаге
        for e, f in wins_combs3:
            num = 10**len(str(e)) # устанавливаем на сколько нужно сместить запятую, чтобы взять корректный остаток
            if combs % num == e: # нам нужно сравнить остаток от деления с выигрышным окончанием числа
                wins += 1 # фиксируем 1 победу
                total += f # увеличиваем наш выигрыш
                break # завершаем цикл
                    
print("Потрачено:", games, "руб.","Побед:", wins, "Выигрыш:", total, "руб.", "Итого:", total-games, "руб.")
        
p_monte_carlo = wins / games # Вероятность из моделирования                          
p = 5 / 1000 + 9 / 1000 + 9 / 1000 + 81 /1000 + 81 /1000 # Теоретическая вероятность

if total-games < 0:
    print("Игра не выгодна игроку")
else:
    print("Игра выгодна игроку")

print("Вероятность из моделирования:", p_monte_carlo)
print("Теоретическая вероятность:", p)
print("Средний выигрыш за все игры:", round(total/games, 2), "руб.", "Средний выигрыш при победе:", round(total/wins, 2), "руб.",)
```
